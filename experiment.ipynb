```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
```

```python
results_dir = Path("results/optimize")

validation_dir = results_dir / "validation"
```

```python
rewards = []
elite_rewards = []
gens = list(range(1, 101, 1))
for i in gens:
    df = pd.read_csv(validation_dir / f"{i}.csv")
    rewards.append(df["reward"].mean())
    elite_rewards.append(df["reward"].iloc[:10].mean())

fig, ax = plt.subplots()
ax.plot(gens, rewards, label="Overall mean")
ax.plot(gens, elite_rewards, label="Elite mean")
ax.set_xlabel("Generation")
ax.set_ylabel("Reward")
ax.legend()
plt.show()
```
