```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
```

```python
results_dir = Path("results/noseed")

validation_dir = results_dir / "validation"
```

```python
rewards = []
elite_rewards = []
best_rewards = []
gens = list(range(1, len(list(results_dir.glob("*.csv"))), 1))
for i in gens:
    df = pd.read_csv(validation_dir / f"{i}.csv")
    rewards.append(df["reward"].mean())
    elite_rewards.append(df["reward"].iloc[:10].mean())
    best_rewards.append(df["reward"].min())

print(rewards)
print(elite_rewards)
print(best_rewards)

fig, ax = plt.subplots()
ax.plot(gens, rewards, label="Overall mean")
ax.plot(gens, elite_rewards, label="Elite mean")
ax.plot(gens, best_rewards, label="Best")
ax.plot
ax.set_xlabel("Generation")
ax.set_ylabel("Reward")
ax.legend()
plt.show()
```
